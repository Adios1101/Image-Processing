Resizing the image using Pillow lib

```python
from PIL import Image
```

```python
img = Image.open('/content/puppy.jpg')
img_resized = img.resize((300, 300))
```

```python
img_resized
```

```python
img_resized.save('puppy_image_resize.jpg')
```

```python
# displaying the image from numpy array

img_res= mpimg.imread('/content/puppy_image_resize.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

Converting RGB images to Grayscale image using openCV

```python
import cv2
```

```python
img = cv2.imread('/content/puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
